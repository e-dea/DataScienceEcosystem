```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizza la distribuzione del valore mediano dell'abitazione
sns.histplot(df['median_house_value'], kde=True)
plt.title('Distribuzione del Valore Mediano dell\'Abitazione')
plt.xlabel('Valore Mediano dell\'Abitazione')
plt.ylabel('Frequenza')
plt.show()
```

tu farai questo esercizio su questi dati: Analizzare la Relazione tra Reddito Mediano e Valore Mediano dell'Abitazione
Potresti voler esplorare come il reddito mediano influisce sul valore mediano dell'abitazione utilizzando un grafico scatter. (ti cerchi cosa vuol dire e lo fai) non usare bot AI prova a fare una ricerca


```python
import numpy as np
import scipy.stats as stats

x = (df.iloc[:, 7])
y = (df.iloc[:, 8])

# Visualizza il grafico a dispersione

plt.scatter(x, y)

# Fit linear regression via least squares with numpy.polyfit
# It returns an slope (b) and intercept (a)
# deg=1 means linear fit (i.e. polynomial of degree 1)
b, a = np.polyfit(x, y, deg=1)
#plt.plot(x, y, '.')
# Plot regression line
plt.plot(x, a + b * x, color="k", lw=2.5);
#creare e plottare coefficiente di correlazione e p value
r, p = stats.pearsonr(x, y)
plt.annotate('r = {:.2f}'.format(r), xy=(0.7, 0.9), xycoords='axes fraction')
plt.show()

print("%f" % p)
```

L'associazione tra le due variabili è significativa :)
